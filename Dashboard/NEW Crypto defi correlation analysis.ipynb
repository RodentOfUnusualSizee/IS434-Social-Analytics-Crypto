```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

aave = pd.read_csv('aave_price_data.csv')
compound = pd.read_csv('compound_price_data.csv')
curve = pd.read_csv('curve_price_data.csv')
maker = pd.read_csv('maker_price_data.csv')
sushi = pd.read_csv('sushi_price_data.csv')
uniswap = pd.read_csv('uniswap_price_data.csv')
```
